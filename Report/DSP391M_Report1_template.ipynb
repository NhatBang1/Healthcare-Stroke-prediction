## I. TỔNG QUAN DỰ ÁN (PROJECT INTRODUCTION)

### 1.1. Đề tài lựa chọn
**Tên đề tài:** Tích hợp mô hình Học máy đa nguồn dữ liệu dự báo nguy cơ Đột quỵ với Ước lượng Tiêu hao Năng lượng nhân tạo (An Integrated Multi-Source Machine Learning Framework for Stroke Risk Prediction with Synthetic Energy Expenditure Estimation)

### 1.2. Lý do chọn đề tài & Động lực (Problem Statement & Motivation)
- Đột quỵ (Stroke) là một biến cố mạch máu não cực kỳ nguy hiểm, là nguyên nhân gây tử vong đứng thứ hai toàn cầu (chiếm khoảng 11% số ca tử vong) và là nguyên nhân chính dẫn đến tàn tật lâu dài ở người lớn. Phát hiện sớm đối tượng có nguy cơ cao giúp can thiệp kịp thời, cải thiện chất lượng sống và giảm chi phí y tế.
- Hiện nay, việc dự báo nguy cơ đột quỵ bằng học máy đối mặt với 2 thách thức lớn:
  1. **Rò rỉ dữ liệu (Data Leakage):** Nhiều bộ dữ liệu y tế chứa thông tin trùng lặp của cùng một bệnh nhân qua nhiều lần thăm khám (multi-visit). Việc phân tách dữ liệu ngẫu nhiên ở cấp độ dòng (row-level) khiến thông tin của cùng một bệnh nhân xuất hiện ở cả tập huấn luyện (train) và kiểm thử (test), tạo nên chỉ số chính xác giả tạo (artificially optimistic accuracy) nhưng hoạt động kém trên dữ liệu thực tế.
  2. **Thiếu hụt chỉ số hành vi (Behavioral Biomarkers):** Mức độ hoạt động thể chất và lượng năng lượng tiêu thụ (calories burned) là những chỉ báo quan trọng đối với nguy cơ tim mạch và đột quỵ. Tuy nhiên, các chỉ số này rất hiếm khi được ghi nhận trong các bệnh án điện tử (EHR) lâm sàng thông thường.
- **Động lực dự án:** Nhóm quyết định phát triển một framework học máy tích hợp kết nối hai tập dữ liệu (lâm sàng và thể chất). Dự án sử dụng giải pháp khử trùng lặp bệnh nhân để ngăn rò rỉ dữ liệu chéo, đồng thời xây dựng một mô hình hồi quy (XGBoost Regressor) để tạo đặc trưng calo tiêu thụ nhân tạo (`Estimated_calories`) từ tuổi, giới tính và BMI, làm giàu dữ liệu y tế để tối ưu hóa dự báo đột quỵ.

### 1.3. Mục tiêu & Phạm vi (Objectives & Scope)
- **Mục tiêu:**
  - Thiết lập một quy trình xử lý dữ liệu sạch, không rò rỉ thông tin bệnh nhân chéo thông qua thuật toán Deduplication.
  - Huấn luyện mô hình hồi quy ước lượng calo (`XGBRegressor`) đạt sai số tuyệt đối trung bình (MAE) thấp để tổng hợp đặc trưng hành vi.
  - So sánh 6 thuật toán học máy phân loại nguy cơ đột quỵ, tối ưu hóa ngưỡng quyết định F1-score trên tập validation để đạt chỉ số Recall tối ưu (tránh bỏ sót ca bệnh).
  - Phân tích đóng góp đặc trưng bằng SHAP để đảm bảo tính giải thích lâm sàng của mô hình.
  - Đóng gói toàn bộ mô hình và pipeline tiền xử lý thành mã nguồn C++ thuần túy (`stroke_predictor_model.h`) phục vụ triển khai thời gian thực trên các thiết bị nhúng biên lâm sàng.
- **Input/Output:**
  - **Đầu vào (Input):** Chỉ số lâm sàng bệnh nhân gồm Tuổi, Giới tính, Chỉ số khối cơ thể (BMI), Huyết áp (RestingBP), Cholesterol (heart_cholesterol), Mức đường huyết (avg_glucose_level), Tiền sử bệnh tim/huyết áp cao, Trạng thái hút thuốc.
  - **Đầu ra (Output):** Nhãn dự báo đột quỵ (0: Không có nguy cơ cao, 1: Có nguy cơ đột quỵ) và xác suất rủi ro đột quỵ chi tiết.

### 1.4. Đóng góp dự kiến
- **Đóng góp kỹ thuật:**
  - Giải quyết bài toán rò rỉ dữ liệu y tế bằng phương pháp gom cụm và khử trùng lặp cấp bệnh nhân (Patient-level Deduplication).

Trong y khoa, tính giải thích là bắt buộc; SHAP giúp nhóm làm sáng tỏ cách mô hình đưa ra dự báo, đồng thời giúp phát hiện các đặc trưng có mối quan hệ phản khoa học (Red-flag features) để loại bỏ.
